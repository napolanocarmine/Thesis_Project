```python
class custom_dict_nGmap(dict_nGmap):
    def __init__(self, *args):
        super(custom_dict_nGmap, self).__init__(*args)

    def is_insertable(self, gm, c, i, assoc):

        """
            Algorithm 60
        """

        """
            if 'c' contains more than one i-cell then return False
        """

        i_cell = set(self.cell_i(i, dart))
        if i_cell > 1:
            return False

        """
            if the i-cell in 'c' is not removable then return False
        """
        for d in i_cell:
            flag = dict_nGmap.is_i_removable(i, d)
            if flag == False:
                return False


        """
            Computing assoc^(-1) that is an associative array between darts, inverse of 'assoc'
        """

        inverse_assoc = ' ' #something 

        for d in assoc:
            if not dict_nGmap.is_i_free(i, d):
                return False
            
            d1 = assoc[d]

            #get the right value of 'n' of dict_nGmap class
            for j in dict_nGmap.n:
                
                if dict_nGmap.alpha[j][d] in assoc and assoc[dict_nGmap.alpha[j][d]] != assoc[d].dict_nGmap.alpha[j]:
                    return False

                if dict_nGmap.alpha[j][d] in inverse_assoc and inverse_assoc[dict_nGmap.alpha[j][d1]] != inverse_assoc[d1].dict_nGmap.alpha[j][d1]:
                    return False


            d2 = dict_nGmap.alpha[i+1][d]

            while d2 not in assoc:
                dict_nGmap.alpha[i+1][dict_nGmap.alpha[i][d2]]

            if assoc[d2] != dict_nGmap.alpha[i][d1]:
                return False

        return True

    

    def insert_nGmap(self, gm, c, i, assoc):
        """
            Algorithm 61
        """

        for d in c.darts:
            dict_nGmap.darts.add[d]
            del c.darts[d]

            if d in assoc:
                dict_nGmap.alpha[i][d] = assoc[d]
                assoc[dict_nGmap.alpha[i][d]] = d

    
    def is_expansible(self, gm, c, i, assoc):

        """
            if 'c' contains more than one i-cell then return False
        """
        i_cell = set(self.cell_i(i, dart))
        if i_cell > 1:
            return False

        """
            if the i-cell in c is not contractible the return False
        """

        for d in i_cell:
            flag = dict_nGmap.is_i_contractible(i, d)
            if flag == False:
                return False

        inverse_assoc = '' #something

        for d in assoc:

            if not dict_nGmap.is_i_free(i, d):
                return False
            
            d1 = assoc[d]

            for j in dict_nGmap.n:
                if dict_nGmap.alpha[j][d] in assoc and assoc[dict_nGmap.alpha[i][d]] != assoc[d][dict_nGmap.alpha[j]]:
                    return False

                if dict_nGmap.alpha[j][d1] in inverse_assoc and inverse_assoc[dict_nGmap.alpha[j][d1]] != inverse_assoc[d1][dict_nGmap.alpha[j]]:
                    return False

            d2 = dict_nGmap.alpha[i-1][d]

            while d2 not in assoc:
                # i think that part is a set_ai
                dict_nGmap.alpha[i].alpha[i-1]

            if assoc[d2] != dict_nGmap.alpha[i][d1]:
                return False

        return True


    
    def expand(self, gm, c, i, assoc):
        """
            Algorithm 63
        """

        for d in c.darts:
            dict_nGmap.create_dart(d)
            del c.darts[d]

            if d in assoc:
                dict_nGmap.alpha[i][d] = assoc[d]
                assoc[d][dict_nGmap.alpha[i]] = d
```
